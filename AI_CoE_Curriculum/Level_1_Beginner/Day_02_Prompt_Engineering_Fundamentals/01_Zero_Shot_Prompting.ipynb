## What is Zero-Shot Prompting?

**Zero-shot** means you give the model a task **without any examples**. You simply describe what you want, and the model uses its pre-trained knowledge to complete it.

This is the simplest form of prompting and works well for:
- Straightforward tasks the model already knows how to do
- When you don't have example data to provide
- Quick prototyping and exploration

## 1. Classification

Ask the model to classify text into categories — no examples needed.

```python
# Sentiment classification
prompt = """
Classify the sentiment of this review as Positive, Negative, or Neutral:

"I go to this restaurant every week, I love it so much."
"""

response = client.models.generate_content(model=MODEL_ID, contents=prompt)
print(f"Sentiment: {response.text.strip()}")
```

```python
# Topic classification
prompt = """
Classify the following news headline into one of these categories: 
Technology, Sports, Politics, Entertainment, Science

Headline: "NASA's James Webb Space Telescope captures stunning images of distant galaxies"
"""

response = client.models.generate_content(model=MODEL_ID, contents=prompt)
print(f"Category: {response.text.strip()}")
```

## 2. Information Extraction

Extract specific information from unstructured text.

```python
# Extract entities
prompt = """
Extract all person names, organizations, and locations from this text:

"Sundar Pichai, CEO of Google, announced a new AI research partnership 
with Stanford University during a conference in San Francisco."
"""

response = client.models.generate_content(model=MODEL_ID, contents=prompt)
Markdown(response.text)
```

## 5. Tips for Better Zero-Shot Prompts

| Tip | Bad Example | Good Example |
|-----|-----------|-------------|
| **Be specific** | "Tell me about dogs" | "List 5 popular dog breeds for families with children" |
| **Specify format** | "Summarize this" | "Summarize this in 3 bullet points" |
| **Set constraints** | "Write about AI" | "Write a 100-word explanation of AI for beginners" |
| **Define the task** | "Fix this" | "Find and fix the syntax error in this Python code" |
