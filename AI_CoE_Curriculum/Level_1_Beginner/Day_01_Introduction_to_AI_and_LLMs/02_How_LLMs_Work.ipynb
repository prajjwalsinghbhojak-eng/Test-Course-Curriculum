## 1. The Transformer Architecture

All modern LLMs are built on the **Transformer** architecture, introduced in the 2017 paper ["Attention Is All You Need"](https://arxiv.org/abs/1706.03762).

### Before Transformers
- Earlier language models used **Recurrent Neural Networks (RNNs)** which processed text **sequentially** (word by word)
- This was slow (couldn't parallelize) and struggled with long-range dependencies

### The Transformer Breakthrough
Transformers process all tokens **in parallel** using a mechanism called **self-attention**. This allows them to:
- Train much faster on large datasets
- Capture long-range dependencies in text
- Scale to billions of parameters

### Simplified Architecture
```
Input Text
    ↓
┌─────────────┐
│ Tokenizer   │  → Converts text to numbers (token IDs)
└─────┬───────┘
      ↓
┌─────────────┐
│ Embedding   │  → Converts token IDs to vectors
│ Layer       │
└─────┬───────┘
      ↓
┌─────────────┐
│ Transformer │  → Multiple layers of self-attention
│ Blocks (×N) │     + feed-forward networks
└─────┬───────┘
      ↓
┌─────────────┐
│ Output      │  → Predicts probability of next token
│ Layer       │
└─────────────┘
    ↓
Generated Text
```

## 2. Self-Attention: The Key Innovation

Self-attention allows the model to look at **all other words** in a sentence when processing any given word.

### Example:
> "The **cat** sat on the mat because **it** was tired."

When the model processes the word "it", attention helps it figure out that "it" refers to "cat" (not "mat"). It does this by computing **attention scores** between every pair of words.

### Multi-Head Attention
Transformers use **multiple attention heads** (typically 8-128) running in parallel. Each head can learn to pay attention to different things:
- One head might focus on **grammar** (subject-verb agreement)
- Another might focus on **semantics** (word meaning relationships)
- Another might focus on **position** (nearby vs distant words)

> 💡 **Key Takeaway:** Self-attention is what makes LLMs so good at understanding context. It lets the model consider the entire input when generating each output token.

## 3. Tokens: The Language of LLMs

LLMs don't work with words — they work with **tokens**. A token is a piece of text that might be:
- A whole word: `hello` → 1 token
- Part of a word: `unhappiness` → `un` + `happiness` → 2 tokens
- A single character: punctuation marks are often individual tokens
- A sub-word: `tokenization` → `token` + `ization` → 2 tokens

### Why Tokens Matter
1. **Cost** — API pricing is per-token (both input and output)
2. **Context limits** — Models have a maximum number of tokens they can process
3. **Performance** — More tokens = slower responses and higher costs

### Rule of Thumb
- **English text:** ~1 token ≈ 4 characters ≈ ¾ of a word
- **1,000 tokens ≈ 750 words**

Let's see tokenization in action:

````python
# Let's explore tokenization with a simple example
# We'll count tokens using a rough approximation first

sample_texts = [
    "Hello world",
    "The quick brown fox jumps over the lazy dog",
    "Artificial Intelligence is transforming industries worldwide",
    "def calculate_average(numbers): return sum(numbers) / len(numbers)",
]

print("Token Estimation (1 token ≈ 4 characters)")
print("=" * 60)
for text in sample_texts:
    char_count = len(text)
    estimated_tokens = char_count / 4  # rough estimate
    word_count = len(text.split())
    print(f"\nText: \"{text}\"")
    print(f"  Characters: {char_count}")
    print(f"  Words: {word_count}")
    print(f"  Estimated tokens: ~{estimated_tokens:.0f}")
````

## 4. Context Windows

The **context window** is the maximum number of tokens the model can process in a single conversation (input + output combined).

### Context Window Sizes

| Model | Context Window | Equivalent |
|-------|---------------|------------|
| GPT-3.5 | 4,096 tokens | ~3,000 words |
| GPT-4 | 128,000 tokens | ~96,000 words |
| Claude 3.5 | 200,000 tokens | ~150,000 words |
| **Gemini 2.5 Pro** | **1,000,000 tokens** | **~750,000 words (a book!)** |

### Why Context Windows Matter
- **Larger windows** = can process more information at once (entire documents, long conversations)
- **Smaller windows** = may lose context in long conversations or when processing large documents
- **Cost tradeoff** = more tokens in the context = higher API costs

> 💡 **Gemini's advantage:** With a 1M token context window, Gemini can process entire codebases or books in a single prompt. This is one of the largest context windows available today.

## 5. How LLMs Generate Text (Next-Token Prediction)

LLMs generate text one token at a time. At each step, the model:

1. Looks at all tokens so far (the prompt + previously generated tokens)
2. Computes a **probability distribution** over all possible next tokens
3. **Samples** one token from that distribution
4. Adds that token to the sequence
5. Repeat until a stop condition is met

### Example:
```
Prompt: "The capital of France is"

Model's predicted probabilities for next token:
  "Paris"  → 92%
  "Lyon"   → 2%
  "a"      → 1%
  "the"    → 0.5%
  ...      → remaining %

Selected token: "Paris" ✓
```

### Key Parameters That Control Generation:

| Parameter | What It Controls | Default |
|-----------|-----------------|--------|
| **Temperature** | Randomness of selection (0 = deterministic, 1+ = creative) | ~0.7-1.0 |
| **Top-K** | Only consider the K most likely tokens | 40 |
| **Top-P** | Only consider tokens whose cumulative probability ≤ P | 0.95 |
| **Max Output Tokens** | Maximum number of tokens to generate | Varies |

> We'll experiment with these parameters hands-on in the next notebook!

## 6. Pre-training vs Fine-tuning vs Prompting

There are three main ways to get an LLM to do what you want:

| Method | Description | Cost | When to Use |
|--------|-----------|------|-------------|
| **Pre-training** | Training from scratch on massive data | 💰💰💰 ($millions) | Only for model companies (Google, OpenAI) |
| **Fine-tuning** | Adapting a pre-trained model on your specific data | 💰💰 ($hundreds-thousands) | When prompting isn't enough (Level 3) |
| **Prompting** | Writing instructions for the model | 💰 (pennies per call) | First approach for everything (Level 1 focus) |

### The Practical Reality
In most real-world applications, **good prompt engineering** gets you 80-90% of the way. You should only consider fine-tuning when:
- You need the model to follow very specific output formats consistently
- You have domain-specific language the model doesn't handle well
- You need to significantly reduce token usage for repeated tasks

> 🎯 **In this course:** Levels 1 and 2 focus on prompting and RAG. Level 3 covers fine-tuning.
