## 1. AI → ML → Deep Learning → Generative AI

Let's break down the hierarchy:

| Term | What It Means | Example |
|------|--------------|--------|
| **Artificial Intelligence (AI)** | Any system that can perform tasks that normally require human intelligence | Chess engines, spam filters, self-driving cars |
| **Machine Learning (ML)** | AI systems that *learn from data* rather than being explicitly programmed | Recommendation systems (Netflix, YouTube) |
| **Deep Learning** | ML using neural networks with many layers | Image recognition, speech-to-text |
| **Generative AI** | Deep Learning models that can *create new content* (text, images, code, audio) | ChatGPT, Gemini, DALL-E, Midjourney |

### Key Insight
Traditional AI **classifies** or **predicts** (e.g., "Is this email spam?"). Generative AI **creates** (e.g., "Write me an email about...").

```
AI (broadest)
 └── Machine Learning
      └── Deep Learning
           └── Generative AI  ← We are here
```

## 2. How Does Generative AI Work? (The 10,000-Foot View)

Generative AI models are trained on massive datasets of text, images, or other content. Through this training, they learn **patterns** — the statistical relationships between words, pixels, or tokens.

When you give a model a prompt, it doesn't "think" or "understand" in the human sense. Instead, it **predicts the most likely next token** based on the patterns it learned during training.

### The Core Process:
1. **Training:** The model reads billions of documents and learns patterns
2. **Prompting:** You give the model an input (a prompt)
3. **Generation:** The model predicts tokens one at a time to create a response

> 💡 **Think of it like autocomplete on steroids.** Your phone's keyboard predicts the next word — LLMs do the same thing, but with much more sophistication and context.

## 5. What is a Large Language Model (LLM)?

An LLM is a type of Generative AI model specifically designed to work with **text**. The "Large" refers to the number of parameters (weights) in the model — modern LLMs have billions to trillions of parameters.

### Key Characteristics of LLMs:

| Property | Description |
|----------|------------|
| **Parameters** | Billions of learned weights (e.g., GPT-4 has ~1.8 trillion) |
| **Training Data** | Trained on large portions of the internet, books, code |
| **Context Window** | How much text the model can "see" at once (e.g., Gemini 2.5 supports up to 1M tokens) |
| **Token-Based** | Models work with tokens, not words (more on this in Notebook 2) |
| **Probabilistic** | Outputs are *sampled* — the same prompt can give different outputs |

### What LLMs Are Good At:
✅ Text generation, summarization, translation  
✅ Question answering and reasoning  
✅ Code generation and debugging  
✅ Following complex instructions  

### What LLMs Struggle With:
❌ Factual accuracy (they can "hallucinate")  
❌ Real-time information (training data has a cutoff)  
❌ Math and precise calculations  
❌ Accessing external systems without tools
