We can use Google's `gemini-2.5-pro` model to evaluate the outputs of other models or your downstream RAG pipelines using native Structured Outputs. This entirely bypasses the need for OpenAI models in evaluation frameworks!

```python
from pydantic import BaseModel
import sys, os
os.environ["COURSE_PASSPHRASE"] = "actionable insights accurate decisions"
repo_root = os.path.abspath(os.path.join(os.getcwd(), "../../.."))
if repo_root not in sys.path:
    sys.path.append(repo_root)
from helpers.auth import get_api_key
from google import genai
import os

client = genai.Client(api_key=get_api_key()))

class EvaluationScore(BaseModel):
    score: int
    rationale: str

prompt = '''
Evaluate the following answer based on the context. Give it a score from 1-10.
Context: The capital of France is Paris.
Answer: The capital is Rome.
'''
response = client.models.generate_content(
    model='gemini-2.5-pro',
    contents=prompt,
    config={'response_mime_type': 'application/json', 'response_schema': EvaluationScore}
)
print(response.text)
```
